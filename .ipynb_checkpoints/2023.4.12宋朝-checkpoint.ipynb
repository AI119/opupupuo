```python
#去重查看有多少官员
df2['姓名'].nunique()
```

```python
#统计每组关系人数量，统计社会关系人数量
aaaa = df2.groupby('姓名').agg({'社会关系人姓名':'count'})
aaaa['社会关系人姓名'].sort_values(ascending = False)
```

## 绘制柱状图


```python
#统计每组关系人数量
aa=df2.groupby('姓名').agg({'社会关系人姓名':'count'})['社会关系人姓名'].sort_values(ascending = False)
```

```python
#绘制社会关系人柱形图
plt.bar(aa.index,aa.values)
plt.title('姓名-社会关系')
plt.xlabel('姓名')
plt.ylabel('社会关系人数')
plt.xticks(rotation=90)
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']
plt.show()
```

```python
#重写x轴内容
x = np.arange(1,189)
y = aa.values
plt.bar(x,y)
plt.title('官员-关系人数')
plt.xlabel('官员')
plt.ylabel('关系人数')
plt.bar
```

```python
#提取出籍贯坐标
df2[["X","Y"]]
```

```python
#统计每个籍贯有多少人，并且以姓名列进行去重
aa = df2.groupby('籍贯').agg({'姓名':'nunique'})
aa
```

```python
"""
第一个方法，按照籍贯分组，提取表格中的X和Y列，使用first方式取出重复的第一列
第二个方法，取出表格中的籍贯、X、Y三列，并按照籍贯进行分组，使用first方式取出重复的第一列
"""

zuobiao = df2.groupby("籍贯")[['X','Y']].first()
# df2[["籍贯","X","Y"]].groupby('籍贯').first()
```

```python
#新增一列，将每个地区的人数添加到新列中
zuobiao['人数'] = aa['姓名']
zuobiao
```

```python
zuobiao[["X","Y"]]
```

```python
#取出每一行的索引作为字典键名，并转换成列表，把每一行的x，y坐标转换成列表，添加到对应的地名中
# zuobiao.index.tolist
address = {}
for city in zuobiao.index.tolist():
    address[city]=zuobiao.loc[city,["X","Y"]].tolist()
address
```

```python
import json as js
js.dumps(address)
```
