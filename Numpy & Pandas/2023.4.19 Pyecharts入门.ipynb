### 引入数据

```python
data1 = pd.read_excel('./data/超市.xlsx')
data1.head()
```

## 处理数据

```python
#data1是超市表格，取出下单年月的数据
data1['下单年'] = data1['订单日期'].dt.year
data1['下单月'] = data1['订单日期'].dt.month
data1['下单年月'] =pd.to_datetime(data1.apply(lambda x:str(x['下单年'])+'-'+str(x['下单月']),axis = 1))

# data1['下单年月'] = pd.to_datetime(data1['下单年月'])
data1['下单年月'] = pd.to_datetime(data1["下单年月"]).apply(lambda x:x.strftime('%Y-%m'))
data1.head()
```

```python
data_1 = data1[['下单年月','销售额']].groupby('下单年月').agg({'销售额':'sum'})
data_1.head()
```

```python
data_1.head()
```

```python
c = (
    Line()
    .add_xaxis(data_1.index.tolist())
    .add_yaxis("商家B",data_1['销售额'].tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title="Line-基本示例"))
    # .render("line_base.html")
)
c.render_notebook()
```

```python
data1.head()
```

```python
data_12 = data1[['下单年月','利润','销售额']].groupby('下单年月').agg({'利润':'sum','销售额':'sum'}).apply(lambda x:round(x*0.0001,2))
data_12.head()
```

```python
c = (
    Line()
    .add_xaxis(data_12.index.tolist())
    .add_yaxis("销售额", data_12['销售额'].tolist(), areastyle_opts=opts.AreaStyleOpts(opacity=0.5),is_smooth=True)
    .add_yaxis("利润", data_12['利润'].tolist(), areastyle_opts=opts.AreaStyleOpts(opacity=0.5),is_smooth=True)
    .set_global_opts(title_opts=opts.TitleOpts(title="Line-面积图"))
    .set_global_opts(legend_opts=opts.LegendOpts(is_show=False),title_opts=opts.TitleOpts(title='销售额走势',pos_left='center'),xaxis_opts=opts.AxisOpts(name_rotate=60,axislabel_opts={"rotate":45}))
    # .render("line_area_style.html")
)
c.render_notebook()
```

## 饼图

```python
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker

c = (
    Pie()
    .add("", [list(z) for z in zip(Faker.choose(), Faker.values())])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
#     .render("pie_base.html")
)
c.render_notebook()
```

```python
# 提取数据
data_pie=data1[['类别','销售额']].groupby('类别').agg({'销售额':'sum'})
data_pie
```

```python
#绘制饼图
x_data = data_pie.index.tolist()
y_data = data_pie['销售额'].tolist()
c = (
    Pie()
    .add("", [list(z) for z in zip(x_data, y_data)])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-不同类别销售额占比分析"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
#     .render("pie_base.html")
)
c.render_notebook()
```

### 绘制圆环图

```python
x_data = data_pie.index.tolist()
y_data = data_pie['销售额'].tolist()
c = (
    Pie()
    .add("", [list(z) for z in zip(x_data, y_data)],radius=['45%','55%'])
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-不同类别销售额占比分析"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
#     .render("pie_ring.html")
)
c.render_notebook()
```

### 玫瑰图

```python
x_data = data_pie.index.tolist()
y_data = data_pie['销售额'].tolist()
c = (
    Pie()
    .add("", [list(z) for z in zip(x_data, y_data)],rosetype='radius',radius='55%')
    .set_global_opts(title_opts=opts.TitleOpts(title="Pie-不同类别销售额占比分析"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
#     .render("pie_rose.html")
)
c.render_notebook()
```

```python
provinceincome_tab = pd.DataFrame(data1.groupby(['省/自治区'])['销售额'].sum()).reset_index()
provinceincome_tab
```

```python
from pyecharts.charts import TreeMap
tree_data = []
sample = {}
for i,j in zip(provinceincome_tab['省/自治区'],provinceincome_tab['销售额']):
    sample = {'value':j,'name':i}
    tree_data.append(sample)
tree_data
```

```python
from pyecharts import options as opts
from pyecharts.charts import TreeMap
c = (
    TreeMap()
    .add("演示数据", tree_data)
    .set_global_opts(title_opts=opts.TitleOpts(title="TreeMap-基本示例"))
#     .render("treemap_base.html")
)
c.render_notebook()
```

### 散点图

```python
data_7 = data1[['子类别','销售额','利润']].groupby('子类别').agg({'销售额':'sum','利润':'sum'})
data_7
```

```python
import pyecharts.options as opts
from pyecharts.charts import Scatter
from pyecharts.commons.utils import JsCode
x_data = data_7['销售额']
y_data = data_7['利润']

c = (
    Scatter()
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(
        series_name="",
        y_axis=y_data,
        symbol_size=20,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_series_opts()
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            type_="value", splitline_opts=opts.SplitLineOpts(is_show=True)
        ),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        tooltip_opts=opts.TooltipOpts(is_show=False),
    )
#     .render("basic_scatter_chart.html")
)
c.render_notebook()
```

```python
from pyecharts.charts import Scatter
from pyecharts.commons.utils import JsCode
bosco_chart = (

    Scatter()
    .add_xaxis(x_data)
    .add_yaxis(
        '',[list(z) for z in zip(data_7['利润'],data_7.index.tolist())],
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True
                                               ,position='top' #将数据标签显示在数据的上方
                                               ,formatter=JsCode("function (params) {return params.value[2];}")))
    .set_global_opts(title_opts=opts.TitleOpts(title='各子类销售额与利润之间关系',pos_left='center',is_show=False),tooltip_opts=opts.TooltipOpts(formatter=JsCode("function (params) {return params.value[2] + ' : ' +params.value[1];}")))
#     .render("basic_scatter_chart111.html")
)
bosco_chart.render_notebook()
```
