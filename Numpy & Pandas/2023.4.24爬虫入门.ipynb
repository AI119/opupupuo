## 实例：从豆瓣电影Top250页面取出文本链接和图像

```python
#打开html文件
with open(r'./data/豆瓣电影 Top 250.html',mode='r',encoding='utf-8') as f:
    lines = f.readlines()
# type(lines)
# # lines
#使用字典保存图片及链接
movie_dict = {
    "标题":[],
    "链接资源":[],
    "图片资源":[]
}
for line in lines:
    #保存标题资源
    if '<span class="title">' in line:
        # print(line)
        title = line.replace('<span class="title">','').replace('\n','').replace('</span>','')
        title = title.strip()
        if 0x4e00 <= ord(title[0]) <= 0x9fa5:
            # print(title)
            movie_dict['标题'].append(title)
    #查看网页结构，进行数据清洗
    elif '<a href="https://movie.douban.com/subject/' in line:
        # print(line)
        movie_herf = line.strip().replace('<a href="','').replace('">','').replace('" class="">','').replace('" class="','')
        if movie_herf not in movie_dict['链接资源']:
            movie_dict['链接资源'].append(movie_herf)
    elif '<img width="100"' in line:
        # print(line)
        img_href = line.split('src="')[1].replace('"','').replace('class=>','').strip()
        movie_dict['图片资源'].append(img_href)
movie_dict
```

```python
#保存电影数据
import csv
titles = movie_dict['标题']
hrefs = movie_dict['链接资源']
imgs = movie_dict['图片资源']
with open(r'./data/电影数据.csv',mode='w',encoding='gbk',newline='') as f:
    csvfile=csv.writer(f)
    csvfile.writerow(['标题','链接资源','图片资源'])
    for i in range(len(title)):
        csvfile.writerow([titles[i],hrefs[i],imgs[i]])
```

```python
import csv
import requests
from fake_useragent import UserAgent
from lxml import etree
# 创建ua自动生成器
ua = UserAgent()
# 伪造的header信息发起请求
faked_header = { 'user-agent':ua.chrome }
# 请求并获取结果
response = requests.request('get','https://movie.douban.com/top250?start=50&filter=', headers=faked_header)
# print(response.headers)
# print(response.text)
# print(type(response.text))
# 把获取的网页信息变成etree对象
root = etree.HTML(response.text)
# 获取电影名
movie_name_list = root.xpath('//div[@class="hd"]//span[@class="title"][1]/text()')
print(movie_name_list)
# print(type(movie_name_list))
# 把数据保存到csv当中
with open('./movie.csv', mode='a', encoding='gbk', newline='') as f:
    csvfile = csv.writer(f)
    csvfile.writerow(['电影名'])
    for i in range(25):
        csvfile.writerow([movie_name_list[i]])
```
