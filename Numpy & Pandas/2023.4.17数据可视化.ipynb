绘制图形

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#设置x，y轴
x = [1,2,3,4,5]
y = [2,4,6,8,10]
#显示图形
plt.plot(x,y)
```

### 引入数据

```python
df1 = pd.read_excel('./data/销售数据.xlsx',index_col='序号')
df1.index
```

按照年月分组

```python
df1['date1'] = pd.to_datetime(df1['创建时间']).apply(lambda x:x.strftime('%Y-%m'))
df1.head()
```

```python
#提取每月的实付金额
data = df1.groupby('date1').agg({'实付金额':'sum'})
data.head()
```

### 绘制折线图

```python
#设定x，y轴数据
x = data.index.tolist()
y = data.values.tolist()
plt.plot(x,y)
plt.show
```

### 设置图片大小

```python
plt.figure(figsize=(20,8),dpi=80)
x = data.index.tolist()
y = data.values.tolist()
plt.plot(x,y)
plt.show()
```

### 设置x轴y轴刻度

```python
#设置画布大小
plt.figure(figsize=(20,8),dpi=80)
x = data.index.tolist()
y = data.values.tolist()
#设置x轴刻度的显示角度
plt.xticks(rotation = 45)
plt.plot(x,y)
plt.show()
```

### 改变绘图的轴刻度线的颜色

```python
#创建一个新的图形窗口
fig = plt.figure()
#在新的图形窗口中创建一个子图，返回一个AxesSubplot对象，111表示讲图形窗口划分成一个行1列1位置为1的子图网格
ax = fig.add_subplot(111)

ax.plot(range(10))
#设置x轴标题和y轴标题
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
#设置颜色
ax.spines['bottom'].set_color('green')
ax.spines['top'].set_color('red')
ax.spines['left'].set_color('pink')
ax.spines['right'].set_color('yellow')
ax.xaxis.label.set_color('blue')
#设置x轴y轴颜色
ax.tick_params(axis = 'x',colors='red')
ax.tick_params(axis = 'y',colors='pink')

plt.show( )
```

### 折点样式

```python
plt.figure(figsize=(20,8),dpi=400)
x = data.index.tolist()
y = data.values.tolist()
#调整x轴显示角度
plt.xticks(rotation = 45)
#导入x和y，通过plot画折线图
plt.plot(x,y,color = 'red', alpha=0.5, linestyle='--',linewidth = 3)
plt.show()
```

修改marker参数来修改折点样式

```python
plt.figure(figsize=(20,8),dpi = 400)
x = data.index.tolist() #x轴位置
y = data.values.tolist() #y轴位置
#调整x轴显示角度
plt.xticks(rotation = 45)
#导入x和y，通过plot画折线图
plt.plot(x,y,color = 'red', alpha=0.5, linestyle='--',linewidth = 3,marker = 'D')
plt.show()
```

图片的保存

```python
plt.figure(figsize=(20,8),dpi = 400)
x = data.index.tolist() #x轴位置
y = data.values.tolist() #y轴位置
#调整x轴显示角度
plt.xticks(rotation = 45)
#导入x和y，通过plot画折线图
plt.plot(x,y,color = 'red', alpha=0.5, linestyle='--',linewidth = 3,marker = 'D')
#保存图片
plt.savefig('./data/t1.png')
plt.show()
```

一图多线

```python
plt.figure(figsize=(20,8),dpi=400)
x = [1,2,3,4,5,6,7]
y = [7,6,5,4,3,2,1]
y1 = [13,15,12,18,14,19,17]
plt.rcParams['font.family'] = "SimHei"
plt.rcParams["axes.unicode_minus"] = False
plt.plot(x,y, color = 'red',label = '自己')
plt.plot(x,y1, color = 'pink', label = '同事')
#显示图解
plt.legend(loc = 0)
plt.show()
```

一图多个子图

```python
x = np.arange(1, 100)
#新建figure对象
fig=plt.figure(figsize=(20,10),dpi=80)
#新建子图1
ax1=fig.add_subplot(2,2,1)
ax1.plot(x, x)
#新建子图2
ax2=fig.add_subplot(2,2,2)
ax2.plot(x, x ** 2)
ax2.grid(color='r', linestyle='--', linewidth=1,alpha=0.3)
#新建子图3
ax3=fig.add_subplot(2,2,3)
ax3.plot(x, np.log(x))
plt.show()
```

```python
x = np.arange(-10,11)
y = x**2
plt.plot(x,y)
#可以调x轴的左右两边
# plt.xlim([-5,5])
#只调整一边
# plt.xlim(xmin=-4)
# plt.xlim(xmin = 4)
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.show()
```

```python
plt.rcParams["axes.unicode_minus"] = False
y = range(0,14,2) # x轴的位置
x = [-3,-2,-1,0,1,2,3]
# plt.figure(figsize=(20,8),dpi=80)
# 获得当前图表的图像
ax = plt.gca()
# 设置图型的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
# 设置底边的移动范围，移动到y轴的0位置,'data':移动轴的位置到交叉轴的指定坐标
ax.spines['bottom'].set_position(('data', 2))
ax.spines['left'].set_position(('data', -3))
plt.plot(x,y)
plt.show()
```

```python
data1=df1.groupby('地区').agg({'实付金额':'sum'})
data1
```

柱形图绘制

```python
x = data1.index
y = data1['实付金额']
plt.bar(x,y)
plt.xticks(range(len(x)),x)
plt.xticks(rotation = -45)
plt.show
```

### 并列条形图

```python
data3=df1.groupby(['地区','sex']).agg({'销量':'sum'})
data3
```

```python
#将上图拆分
data3=df1.groupby(['地区','sex']).agg({'销量':'sum'}).unstack()
data3.head()
```

```python
data3.plot(kind='bar')
plt.title('不同地区不同性别销量对比')
plt.xlabel('地区')
plt.ylabel('销量')
plt.show()
```

3.罗列柱形图(堆叠柱形图)

```python
data3=df1.groupby(['地区','sex'])['销量'].sum().unstack()
data3.plot(kind='bar',stacked=True)
plt.title('不同地区不同性别销量对比')
plt.xlabel('地区')
plt.ylabel('销量')
```

### 饼图

```python
data4=df1[['地区','销量']].groupby('地区').agg({"销量":"sum"})
data4
```

```python
#绘制图表
# 数据
regions =data4.index
sales = data4['销量']
# 饼图
plt.pie(sales, labels=regions, autopct='%1.1f%%')
# 图表标题
plt.title('各地区销量占比')
# 显示图表
plt.show()
```
