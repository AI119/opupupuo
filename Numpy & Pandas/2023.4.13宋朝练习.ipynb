```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'D:\Jupyter notebook\Numpy & Pandas\officer_kinship_song.csv')
df2 = pd.read_csv(r'D:\Jupyter notebook\Numpy & Pandas\officer_relationship_song.csv')
# df2是官员政治关系表 df1是官员亲戚表
df1.query("亲戚姓名=='丁D'")
```

提取官员姓名

```python
#提取官员姓名转换成集合
offer = set(df2['姓名'])
#所有官员姓名
offer
```

提取亲戚姓名

```python
#提取亲戚姓名转换成集合
kin = set(df1['亲戚姓名'])
kin
```

```python
#以姓名进行分组统计每个人的政治联系
df2['联系'].groupby(df2['姓名']).value_counts()
```

```python
a = df2['联系'].groupby(df2['姓名']).value_counts().unstack().copy()
#a是拆分后的表格的拷贝
a
```

```python
#用0填充空值
a.fillna(0,inplace= True)
a
```

```python
#取亲戚姓名和官员姓名的交集
offer&kin
```

```python
#筛选官员亲属为官数

kin_officer=df1.query('亲戚姓名=="曾布" or 亲戚姓名=="曾肇" or  亲戚姓名=="王安石" or 亲戚姓名=="王安礼" or 亲戚姓名=="章綡" or 亲戚姓名=="章縡" or 亲戚姓名=="苏轼" or 亲戚姓名=="苏辙" or 亲戚姓名=="范仲淹" or 亲戚姓名=="范纯仁" or 亲戚姓名=="范纯礼" or 亲戚姓名=="赵祯" or 亲戚姓名=="赵顼" or 亲戚姓名=="韩忠彦" or 亲戚姓名=="韩琦"')
kin_officer.drop_duplicates('亲戚姓名')['亲戚姓名'].groupby(kin_officer['姓名']).count()
```

```python
#对统计出的亲戚姓名进行去重
kin_officer['亲戚姓名'].groupby(kin_officer['姓名']).nunique()
```

```python
#在拆分的列表中添加新列，内容为亲戚为官数
a['亲戚为官数'] = kin_officer['亲戚姓名'].groupby(kin_officer['姓名']).nunique()
a.fillna(0)
a
```
