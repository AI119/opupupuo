## pandas新增数据列

```python
fpath_1 = './data/beijing_tianqi_2018.csv'
df_1 = pd.read_csv(fpath_1)
df_1
```

```python
#将所有的摄氏度替换为数值
df_1.loc[:, 'bWendu'] = df_1['bWendu'].str.replace('℃', '').astype('int32')
```

```python
df_1.loc[:, 'yWendu'] = df_1['yWendu'].str.replace('℃', '').astype('int32')
```

```python
#添加温差列
df_1.loc[:, "温差"] = df_1["bWendu"] - df_1["yWendu"]
df_1
```

## df.apply方法

```python
#使用apply方法，应用自定义函数到列表中
def get_wendu_type(x):
    if x["bWendu"] > 33:
        return '高温'
    if x["yWendu"] < -10:
        return '低温'
    return '常温'


df_1.loc[:, '温度类型'] = df_1.apply(get_wendu_type, axis=1)
df_1
```

### 使用assign方式添加多列

###### assign() 方法的参数是一个函数，这个函数将被应用于数据框的每一行，用于计算新的列的值。

```python
df_1 = df_1.assign(
    yWendu_huashi=lambda x: x['yWendu'] * 9 / 5 + 32,
    bWendu_huashi=lambda x: x['bWendu'] * 9 / 5 + 32
)
df_1
```

### 按条件选择分组分别赋值

```python
df_1['wencha_type'] = ''
df_1.loc[df_1["bWendu"]-df_1["yWendu"]>10,"wencha_type"]="温差大"
df_1.loc[df_1["bWendu"]-df_1["yWendu"]<=10,"wencha_type"]="温度正常"
df_1
```

## 相关系数和协方差

##### 用途（超级厉害）
1. 两只股票，是不是同涨同跌？程度多大？正相关还是负相关？
2. 产品销量的波动，跟哪些因素正相关、负相关，程度有多大？
来自知乎，对于两个变量x、y:
3. 协方差：衡量同向反向程度，如果协方差为正，说明x,y同向变化，协方差越大说明同向程度越高;
如果协方差为负，说明x,y反向运动，协方差越小说明反向程度越高
4. 相关系数： 衡量相似度程度，当他们的相关系数为1时，说明两个变量变化时的正向相似度最大，
当相关系数为-1时，说明两个变量变化的反向相似度最大

```python
#协方差矩阵
df_1.cov()
```

```python
#相关系数矩阵
df.corr()
```

```python
#单独查看某两个变量之间的相关系数
df_1["aqi"].corr(df_1["yWendu"])
```

```python
#空气质量和温差的相关系数
df_1["aqi"].corr(df_1["bWendu"]-df_1["yWendu"])
```

# 缺失值处理
isnull和notnull:检测是否是空值，可用于df和series
dropna:丢弃、删除缺失值

1. axis:删除行还是列,{0 or 'index',1 or 'columns'},default 0


2. how:如果等于any则任何值为空都删除，如果等于all则所有值都为空才删除


3. inplace:如果为True则修改为当前df,否则不回新的df

fillna:填充空值

1. value: 用于填充的值，可以是单个值，或者字典（key是列名，value是值）

2. method=ffill使用前一个不为空的值填充forword fill;method=bfill使用后一个不为空的值填充backword fill


3. axis:按行还是列填充，{0 or 'index',1 or 'columns'}


4. inplace: 如果为True则修改为当前df,否则返回新的df

```python
#删除全是空值的列
studf.dropna(axis="columns",how="all",inplace=True)
```

```python
#删除全是空值的行
studf.dropna(axis=0,how="all",inplace=True)
studf
```

## 数据填充
fillna:填充空值

1. value: 用于填充的值，可以是单个值，或者字典（key是列名，value是值）

2. method=ffill使用前一个不为空的值填充forword fill ;    method=bfill使用后一个不为空的值填充backword fill


3. axis:按行还是列填充，{0 or 'index',1 or 'columns'}


4. inplace: 如果为True则修改为当前df,否则返回新的df

```python
#使用前一个不为空的值填充空值
studf.loc[:,"姓名"]= studf['姓名'].fillna(method="ffill")
studf
```

```python
#使用固定值填充
studf.loc[:,"分数"]=studf["分数"].fillna('缺考')
studf
```

## 排序
Series的排序：
Series.sort_values(ascending=True,inplace=False)
参数说明：
ascending:默认为True升序排序，为False降序排序
inplace:是否修改原始Series
DataFrame的排序：
DataFrame.sort_values(by,ascending=True,inplace=False)
参数说明：
by:字符串或者List<字符串>，单列排序或者多列排序
ascending:bool或者List,升序还是降序，如果是list对应by的多列
inplace:是否修改原始DataFrame
